```python
# emp, dept 테이블
# 테이블 간의 결합 -> 사원의 정보 테이블과 부서의 정보 테이블 -> 조건에 맞게 결합이 되는 형태 -> 조인 결합
# 조인 결합 -> pandas에서는 merge() 함수를 이용
```

- 유니언 결합
    - 단순 결합 ( 행, 열 )
    - pandas에서는 concat([데이터프레임,데이터프레임, ...]) 함수를 이용
    - 단순 결합의 형태이기 때문에 데이터프레임의 개수의 제한 x
        - 매개변수
            - axis
                - 행으로 결합할 것인가, 열로 결합할 것인가
            - ignore_index
                - 결합되는 축을 초기화 할것인가
    - SQL에서는 UNION 키워드를 이용해서 결합
        - SELECT문 UNION SELECT문
- 조인 결합
    - 특정한 조건에 맞춰서 열을 결합
    - 데이터프레임과 데이터프레임간의 결합 -> 특정 컬럼의 데이터들이 같은 값들로 이루어져있는 경우 열을 추가하는 방식
    - pandas, DataFrame에 내장된 merge() 함수를 이용
    - 조건에 맞춰서 결합이 되기 때문에 데이터프레임은 2개 결합
        - 매개변수
            - on
                - 조건식 -> 2개의 데이터프레임이 모두 가지고 있는 컬럼의 이름
                - 각 데이터프레임의 컬럼의 이름이 다른 경우에는 left_on, right_on 매개변수를 이용\
            - how
                - 조인 결합의 방식(left, right, inner, outer)
    - SQL에서는 JOIN 키워드를 이용해서 결합
        - select * from table1 inner join table2 on table1.col = tabl2.col

```python
# where(인덱스의 조건 - 필터), group by(그룹화), having(인덱스의 조건 - 필터), order by(정렬)
# 1. where(그룹화를 하기 전에 데이터를 필터)
# 2. group by(그룹화)
# 3. having(그룹화된 데이터에서 필터)
# 4. order by(최종 결과를 이용하여 정렬을 변경)
# 직종에서 ANALYST을 제외하고 직종별로 그룹화하여 SAL의 평균을 구한 뒤,
# 평균의 값이 2500이상인 데이터에서 평균 sal이 높을 순서대로 출력

query_17 = """
    select job, avg(sal) as mean from emp
    where job != 'ANALYST'
    group by job
    having mean >= 2500
    order by mean desc
"""
sql(query_17)
```
