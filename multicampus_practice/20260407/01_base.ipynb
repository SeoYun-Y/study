```python
# 조건문(if문)
# if 뒤의 조건식에 따라 참인 경우에 실행하는 코드와
# 거짓인 경우 실행하는 코드를 나눠서 작성
# python -> if 조건식 :
# 일반적인 언어 -> if (조건식) {  }
a = 11
# if 문의 시작
if a < 10 :
    # 조건식이 참인 경우 실행할 코드를 작성
    print('a는 10보다 작다')
# 거짓인 경우 실행할 코드
else:
    print('a는 10보다 크거나 같다')
# if ~ else 문에서는 하나의 지역의 코드만 실행이 된다.
```

```python
# 조건문에서 조건식이 여러개로 나눠지는 경우
score = 73 # C 학점

# 1번 조건식 : score가 90점 이상이라면
if score >= 90:
    # score가 90점 이상인 경우에만 실행
    print('A학점')
else:
    # score가 90점 미만인 경우 실행되는 지역
    if score >= 80:
        # score가 90점 미만이고 80점 이상인 경우 실행
        print('B학점')
    else:
        # score가 80점 미만인 경우
        print('C학점')
```

```python
# if ~ else를 사용 -> if 뒤의 조건식의 개수를 변경
if score >= 90:
    print('A학점')
# B학점의 조건식 : score가 90점 미만이고(조건식1) 80점 이상(조건식 2)
if (score < 90) and (score >= 80):
    print('B학점')
# C학점의 조건식 :
# score가 80점 미만 그리고 70점 이상
if (score < 80) and (score >= 70):
    print('C학점')
# F학점의 조건식:
# score가 70점 미만
if score < 70:
    print('F학점')
```

```python
# if ~ elif ~ else문
# input() -> 코드를 실행할 때 입력값을 받아온다( 문자형으로 데이터를 받아온다 )
score = int(input('점수를 입력하시오: '))

if score >= 90:
    print('A학점')
elif score >= 80:
    print('B학점')
elif score >= 70:
    print('C학점')
else:
    print('F학점')

print('프로그램 종료')
```

```python
# 반복문
# 기본형 : for 변수명 in range(시작값, 종료값, 증가값)
# for -> 반복하겠다
# 변수명 -> 반복실행하면서 데이터가 할당되는 공간
# in -> ~안에 -> in 뒤에 자료형 데이터들 안에 데이터들을 사용한다.
# range() -> 시작값부터 종료값까지 증가값을 더해서 자료형 데이터를 구성
print(
    list(range(1,10,1))
)

# for i in range(1,10,1)
# for i in [1,2,3,4,5,6,7,8,9] -> 둘이 같은 의미
```

```python
# 반복횟수만 필요한 경우 -> i의 할당되는 데이터를 굳이 사용 안하는 경우
# index의 값으로 사용하는 경우
for i in range(1,10,1):
    print(i)
```

```python
# 자료형 데이터를 반복문에서 사용하는 경우
for i in [1,2,3,4,5,6,7,8,9]:
    print(i)
```

```python
# range() 함수에서 시작값과 증가값은 기본값이 지정
# 시작값은  0 기본값, 증가값은 1이 기본값
print(list(range(10)))
print(list(range(0,10,1)))
```

```python
# 1~10까지의 합계
# 누적합을 어딘가에 저장 -> 공간을 생성하여 0을 대입
result = 0

for i in range(1,11,1):
    # i에는 1 대입 -> 반복 실행할때마다 1씩 증가
    # result = result + i
    result += i
    # print(result) -> print() 반복문 안에 존재 누적합이 될때마다 출력이 실행
print(result) # --> 반복문이 끝나고 print()가 실행 -> 단 한번만 실행
```

```python
result = 0
for i in range(10):
    # i에는 0대입 -> 반복 실행때마다 1씩 증가(0~9)
    i += 1
    result += i
print(result)
```

```python
# while문

# 시작값 지정
i = 0
# while 조건식: 조건식이 거짓이 될때까지 반복 실행
# 조건식이 거짓이 될 수 있도록 반복 실행 코드에 시작값을 증가/감소
# break문을 이용해서 반복문을 강제 종료
# 시작값으로 조건식을 사용했을때 조건식은 참
while i < 10:
    # 초기값 i를 1씩 증가
    i = i + 1
    print(i)
```
