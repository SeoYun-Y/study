- 변수 : 데이터가 저장되는 공간에 주소 대신에 사용할 별칭

- 불변 데이터
    - 문자, 숫자, 튜플
    - 각각의 다른 공간에 데이터가 대입
    - a = 10 / b = a --> a = 10, b = 10 -> b에 1을 더한다 -> a는 10 그대로 유지
    - 함수에서 매개변수의 기본값
        - 불변 데이터는 매번 데이터가 다른 값들이 대입
        - 휘발성 데이터로 유지

- 가변 데이터
    - 리스트, 딕셔너리
    - copy() 함수를 이용하지 않으면 같은 위치를 지정
    - list_a = [1,2,3] / list_b = list_a / list_b에 원소를 추가하면 list_a도 원소가 추가
    - 함수에서 매개변수의 기본값
        - 값을 변경하지 않으면 해당 데이터는 유지하는 성향
        - def func(a, b= [1,2,3]): b.append(a) return b -> func(1) [1,2,3,1] -> func안에 b라는 지역변수는 [1,2,3,1]을 유지
        - 가변데이터를 기본값으로 사용하는 경우는 copy() 함수를 이용

- 조건문
    - if 뒤의 조건식이 참이라면 코드를 실행
    - else 문을 사용하면 조건식이 거짓인 경우 코드를 실행
    - if (조건식) 기본형인데 조건식 부분에 다른 데이터가 대입이 되는 경우
        - 강제적으로 bool 형으로 데이터가 전환
        - 데이터의 존재 여부로 T / F 판단
    - 여러 개의 조건식이 존재한다면 if ~ elif ~ else문 사용

- 반복문
    - for
        - for 변수명 in range(시작값, 종료값, 증가값)
            - 시작값부터 종료값 전까지 증가값만큼 값을 변화시키면서 반복 실행
        - for 변수명 in 자료형데이터(list,dict,tuple,Series)
            - 자료형 데이터의 각 원소들을 대입하면서 반복 실행
            - 2차원 데이터가 들어오면 [[1,2,3],[4,5,6],[7,8,9]]
                - 3번 반복 ( len(자료형데이터) )
                - [1,2,3] 대입 -> [4,5,6] 대입 -> [7,8,9] 대입
    - while
        - 조건식이 거짓이 될때까지 반복 실행하는 반복문
        - 조건이 거짓이 될 수 있도록 특정 값을 변화 ( i += 1 )
        - 조건이 거짓이 되지 않으면 -> 무한 반복
        - 반복횟수가 명확하지 않을때는 for문보다는 while문이 더 적합할 수 있다
    - break
        - 반복문을 강제로 종료
    - continue
        - 반복문으로 되돌아간다.(continue 밑의 코드는 실행x)

- 함수
    - 반복적으로 사용이 되는 코드들을 묶어서 하나의 행동(연산) 지정
    - 입력값을 받아서 추가적인 연산을 하고 결과를 되돌려준다.
    - 사용을 안해도 되지만 사용을 권장하는 이유는 -> 가독성이 올라가고 코드의 에러를 줄여준다.

- 클래스
    - 붕어빵 틀!
    - 변수 + 함수의 결합체
    - 독립적인 변수를 저장하는 형태로 그 안에서 함수를 이용해서 데이터들을 변화
    - 같은 객체(class)들끼리 서로 공유가 가능한 변수 영역( 클래스 변수 )

- 모듈, 라이브러리, 프레임워크
    - 코드를 분할 관리하기 위해
        - 가독성 올라간다
        - 유지보수가 편해진다.
    - 배포의 중요성
        - 사용자가 모든 코드를 작성하지 않고 모듈을 로드함으로써 기능을 사용
    - .py 파일로 이루어져있다

```python
def func_1(a,b=[]):
    result = b.copy()
    result.append(a)
    return result
```

```python
func_1(10)
```

```python
# 클래스변수가 불변데이터인 경우 : 객체들끼리 공유를 시키는법
class Class_1:
    text = 'Hello'

    def __init__(self,a,b):
        self.a = a
        self.b = b
        self.text = Class_1.text    # 불변 데이터의 =의 의미는 다른 공간에 데이터를 저장 ( 독립 데이터 생성 )
```

```python
class1= Class_1(10,20)
class2= Class_1(50,60)
```

```python
# 클래스 변수의 데이터를 변경해보자( 독립 공간의 데이터를 변화 )
class1.text = 'Hi'
```

```python
print(class1.text,class2.text)
```

```python
class Class_2:
    list_a = []

    def __init__(self,a):
        self.a = a
        self.b = Class_2.list_a
        # 가변데이터에서 =은 얕은 복사 -> 같은 주소를 공유한다
```

```python
class3 = Class_2(10)
class4 = Class_2(30)
```

```python
# 클래스 변수에 50을 추가 -> 가변 데이터는 공유가 가능
class3.b.append(50)
```

```python
print(class3.b,class4.b)
```

```python
# 불변 데이터가 클래스 변수일때 공유를 시켜보자
class Class_3:
    text = 'Hello'
    
    def __init__(self,a):
        self.a = a

    # 클래스 변수의 데이터를 변경하는 함수
    def func_1(self,_text):
        Class_3.text = _text
        # 접근 방식이 Class_3안에 text에 데이터를 변경
```

```python
class5 = Class_3(10)
class6 = Class_3(40)
```

```python
print(class5.text,class6.text)
```

```python
# func_1() 함수를 이용해서 클래스 변수를 변경
class5.func_1('Hi')
```

```python
print(class5.text)
```

```python
print(class6.text)
```
