```python
def func(x, acc = []):
    # 2개의 매개변수가 존재
    # acc 빈 리스트는 가변의 데이터 -> 기본값을 그대로 사용하면 데이터들을 유지 ( 주소값 )
    acc.append(x)
    return acc
print(func(1))
print(func(2))
print(func(3, []))
```

```python
# class에서 문자열 클래스 변수 변경
# @classmethod -> 데코레이터 함수
class Test:
    text = 'Hello'

    def __init__(self,a):
        self.a = a
    def func_1(self,x):
        Test.text = x
```

```python
class_1 = Test(10)
class_2 = Test(20)
```

```python
print(class_1.text)
print(class_2.text)
```

```python
# class 변수 text를 변경하는 함수를 호출
class_1.func_1('Hi')
```

```python
print(class_1.text)
print(class_2.text)
```

```python
class Test2:
    text = 'Hello'
    
    # 일반적인 class 내부의 함수에서 첫번째 매개변수에는 생성이 된 class의 주소값이 입력(자동)
    def __init__(self,a):
        self.a = a

    # classmethod : 클래스의 이름에 접근할 수 있는 데코레이터 ->
    # 함수에는 첫번째 매개변수에 클래스의 이름이 입력(자동)
    @classmethod
    def func_1(cls,a):
        # classmethod를 이용한 함수는 첫번째 매개변수명을 cls
        # cls에는 해당 class의 이름이 인자로 사용
        # 수정 중에 class 이름이 변하더라도 해당 cls는 그대로 유지 가능
        cls.text = x
```

```python
class_3 = Test2(10)
class_4 = Test2(20)
```

```python
print(class_3.text)
print(class_4.text)
```
