### 문자열 내장 함수

- upper() : 문자열을 모두 대문자로 변환
- lower() : 문자열을 모두 소문자로 변환
- swapcase() : 대문자를 소문자로 소문자를 대문자로 변환
- title() : 단어의 첫글자만 대문자로 변환
- strip() : 기본적으로는 문자열의 시작과 끝에 존재하는 공백을 제거
    - lstrip() : 문자열 시작의 공백을 제거
    - rstrip() : 문자열 끝의 공백을 제거
- replace() : 특정한 문자를 찾아서 해당 문자를 다른 문자로 치환
    - 문자 사이에 특정 문자를 치환 ( strip() 함수는 문자의 시작과 끝만 확인하고 replace()는 문자를 모두 확인 )
- split() : 특정 문자를 찾아서 문자를 나눠주는 기능 -> 입력값은 문자, 출력값은 list

```python
# sort() -> 오름차순 / 내림차순 정렬을 할 수 있는 새로운 함수
# 오름차순 정렬 -> min()
# 내림차순 정렬 -> max()

# 함수 선언
    # 매개변수 2개
        # values 인자의 개수가 가변인 형태
        # reverse 기본값이 False(오름차순 정렬) // True로 변경한다면 ( 내림차순 정렬 )

def custon_sort(*values, reverse = False):
    # 정렬이 되는 데이터들을 저장할 공간(빈 리스트 생성)
    result = []
    # values의 데이터가 수정을 할 수 있도록 list로 변환
    values = list(values)
    # values의 길이만큼 반복하는 반복문 생성 : values의 type이 tuple이기 때문에
    for i in range(len(values)):
        # reverse의 값이 True라면 max() 함수를 이용해서 values의 최댓값을 추출
        if reverse:
            value = max(values)
        else:
            value = min(values)
        result.append(value)
        values.remove(value)
    return result
```

```python
custon_sort(3,2,4,1, reverse = True)
```

#### 연습문제
- 매개변수 2개
    - datas : 인자의 개수가 가변인 매개변수
    - types : 기본값의 설정 ('max')
- types가 max라면 datas들 중 가장 큰 값을 되돌려준다.
- types가 min이라면 datas들 중 가장 작은 값을 되돌려준다.
- types가 mean이라면 datas들의 평균 값을 되돌려준다.
- types가 위의 3가지 경우가 아니라면 'types 데이터 입력값이 잘못되었습니다' 라는 출력

```python
def func_12(*datas,types = 'max'):
    if types == 'max':
        return max(datas)
    elif types == 'min':
        return min(datas)
    elif types == 'mean':
        res_sum = sum(datas)
        result = res_sum/len(datas)
    else:
        print("types 데이터 입력값이 잘못되었습니다.")
        result = ''
    return result
```

```python
print(func_12(1,5,7))
print(func_12(1,4,7,types = 'min'))
print(func_12(1,4,7,types='mean'))
print(func_12(1,4,7,types='sub'))
```
