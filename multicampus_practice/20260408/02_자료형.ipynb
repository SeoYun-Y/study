### 자료형 데이터
- 여러개의 데이터를 하나의 변수에 대입하기 위한 형태

- tuple
    - 자료형 데이터 중에 가장 기본적인 형태( 가장 메모리를 덜 먹는 데이터의 형태 )
    - ()를 이용해서 데이터를 묶어서 표현
    - 한번 생성이 되면 수정 불가능 ( 원소를 수정하거나 추가 삭제 작업을 할 때는 새로운 tuple를 생성)
    - list
        - []를 이용해서 데이터를 묶어서 표현(생성)
        - tuple과는 다르게 각 원소들을 수정하거나 추가, 삭제 가능(리스트에서 제공하는 내장 함수를 이용)
        - 원소의 개수를 0개부터 생성이 가능( 비어있는 리스트를 생성 -> 여러 개의 데이터를 하나의 리스트에 추가하여 새로운 데이터의 묶음 생성 )
    - dict
        - {}를 이용해서 데이터를 묶어서 표현(생성)
        - key : value의 형태로 데이터가 구성( list에서 index를 key로 표현한 방식 )
        - key를 기준으로 데이터를 선택, 수정, 삭제 등이 가능
        - 원소의 개수를 0개부터 생성이 가능

```python
# list에서 sort() 함수 -> 리스트의 각각의 원소들을 비교하여 가장 작은 데이터를 처음으로
# 가장 큰 데이터를 마지막에 배치하여 오름 차순 정렬하는 함수
sample_list = [30,10,50,7]
# 가장 작은 숫자부터 대입이 될 빈 리스트가 생성
sort_list = []

# sample_list의 길이만큼 반복 실행하는 반복문을 생성
for i in range(len(sample_list)):
    # 기준이 되는 list의 첫번쨰 원소를 변수에 저장 ( 첫번째 데이터가 가장 작은 데이터라 생각하고 대입)
    min_value = sample_list[0]
    for j in range(1, len(sample_list)):
    # 2번째 데이터부터 min_value와 비교
    # min_value(30)과 sample_list[1] -> 10를 비교 --> min_value는 10대입
    # min_value -> 10과 sample_list[2] -> 50을 비교 --> min_value 유지(데이터의 변화x)
    # min_value -> 10과 sample_list[3] -> 7을 비교 --> min_value 7 대입
    # j는 sample_list의 인덱스 (두번째 데이터)
    # min_value와 크기 비교
        if min_value > sample_list[j]:
            # min_value를 sample_list[j]으로 수정
            min_value = sample_list[j]

    # sort_list에 min_value 추가
    # print(min_value)
    # break
    sort_list.append(min_value)
    # sample_list에서 7을 제거 (remove())
    sample_list.remove(min_value)
    print(sample_list)
    print(sort_list)
```

```python
sample_list = [30,10,50,7]
sort_list = []

# min() : 여러개의 데이터 중 가장 작은 값을 반환
# sample_list에서 가장 작은 값을 추출 -> sort_list에 추가 -> sample_list에서 제거
# sample_list의 길이만큼 반복 실행
for i in range(len(sample_list)):
    # min() 함수를 이용하여 sample_list에서 가장 작은 데이터를 추출하여 변수에 저장
    min_value = min(sample_list)
    # print(min_value)
    # break
    # sort_list에 추가
    sort_list.append(min_value)
    # sample_list.remove(min_value)
print(sort_list)
```

```python
# 내림차순 정렬 -> 리스트에서 가장 큰 데이터를 추출
# max() : 여러개의 데이터에서 가장 큰 값을 반환
sample_list = [30,10,50,7]
sort_list = []

for i in range(len(sample_list)):
    max_value = max(sample_list)
    sort_list.append(max_value)
    sample_list.remove(max_value)
print(sort_list)
```
