```python
# break문
# 반복문을 강제적으로 종료
for i in range(10):
    # 10번 반복을 하는 반복문 (i는 0부터 대입되서 9까지)
    # i가 3일때(조건문) 반복문을 종료
    if i == 3:
        break
    print(i)
print('-'*16)
print(i)
```

```python
# while + break
# 1부터 1씩 증가시켜서 누적합이 1500000이 넘어가는 시점은 언제이며 누적합은 몇인가?

# 누적합 -> 값들이 더해질 공간이 필요
result = 0
# while문에서 시작값을 지정
i = 1

# while 조건식( bool ) -> 반복횟수가 명확하지 않을때 -> 무한 루프( while True ) ->
# 조건(누적합이 1500000보다 크거나 같으면)이 충족되면 반복문을 강제 종료

while True:
    # result에 i의 값을 더한다.
    result += i
    if result >= 1500000:
        break
    # 시작값을 변화( 1씩 증가 )
    i += 1
# i의 값을 출력 ( 1부터 i까지 반복했다는 출력 값 )
print(i)
# result 값을 출력 ( 1부터 i까지의 누적합 )
print(result)
```

```python
# 누적합이 8400은 넘어가는 i의 값과 누적합의 값을 for문으로 확인

result = 0

for i in range(1001):
    result += i
    if result >= 8400:
        break
print(i)
print(result)
```

```python
# case1  (range(2,101,2)) --> 효율적인 방법 (반복 횟수가 반으로 줄어듬)

result = 0
for i in range(2,101,2):
    result += i
print(result)
```

```python
# case2 (반복문 안에 짝수의 조건식을 이용하여 조건식이 참이라면 result에 더해준다)
# 1부터 100까지의 짝수의 합

result = 0

for i in range(1,101):
    if(i % 2 == 0):
        result += i
print(result)
```

```python
# continue -> 아래의 코드를 전부 무시하고 반복문으로 되돌아간다.
for i in range(10):
    continue # print(i)줄로 가지 못함
    print(i)
```

```python
for i in range(10):
    if i < 3:
        continue
    print(i)
```

```python
# case3: 홀수인 경우 continue를 이용하여 반복으로 되돌아간다
result = 0

for i in range(1,101):
    if i % 2 != 0:
        continue
    result += i
print(result)
```

```python
# 다중 반복문: 반복문 안에 반복문이 존재( 2차원 데이터에서 종종 활용 )

# 첫번째 반복문의 시작
for i in range(5):
    # 두번째 반복문이 시작( 첫번째 반복문에서 사용하는 변수 외의 변수명을 사용 )
    # 5번 반복
    for j in range(5):
        # print('i='+ str(i) + " j=" + str(j))
        # print('i=',i ,'j=', j)
        # f-string : 문자열 사이에 변수의 값을 대입하여 하나의 텍스트 생성
        # f"문자{변수명}"
        print(f'i={i} j={j}')
# 최종 반복 횟수는 25회
```

```python
# 2개의 주사위가 존재
# 2개의 주사위의 합이 5의 배수인 모든 경우를 출력하시오.
# (1,4),(2,3),(3,2),(4,1) -> 합이 5인 경우
# (4,6),(5,5),(6,4) -> 합이 10인 경우
# 총 경우의 수는 -> 36개

# 첫번째 주사위의 경우의 수 반복문으로 생성
for i in range(1,7):
    # 두번째 주사위의 경우의 수 반복문으로 생성
    for j in range(1,7):
        # 두 개의 주사위의 합이 5의 배수라면? (조건식 생성)
        if (i+j) % 5 == 0:
            # 조건식이 참인 경우 두 개의 주사위의 숫자를 출력
            print(f'i={i} j={j}')
```
